```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("titanic")
```

```python
data.head()
```

```python
# histogram of passengers class count
sns.histplot(data["pclass"])
plt.show()
```

```python
# Bar chart for the 'Sex' variable
sns.countplot(x='sex', data= data)
plt.title('Distribution of Genders on the Titanic')
plt.show()
```

```python
# creating histogram
# Histogram for the 'Age' variable
plt.figure(figsize=(10, 6))
sns.histplot(data['age'])
plt.title('Distribution of Ages on the Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
